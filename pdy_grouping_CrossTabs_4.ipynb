# Cross tabs

* `syntax:` pd.crosstab(index=[col/s], values = [col/s], func/s)

```python
df.info()
```

```python
pd.crosstab(index = df['region'],
            columns = df['sales_channel'],
            values = df['total_cost'],
            aggfunc = [np.min,np.max,np.mean])
```

```python
dfdt.drop(columns = ['Cabin'], inplace = True)
```

```python
pd.crosstab(index = [dfdt['Embarked'],dfdt['Pclass']],
            columns = dfdt['Survived'],
            values = dfdt['Age'],
            aggfunc = [np.min,np.max,np.mean])
```
