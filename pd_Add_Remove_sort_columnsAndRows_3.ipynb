# Add and remove of rows and columns

## `Adding and Removing columns:` it is basically same thing that we did updating values we can create a column and pass series of values that column to hava

```python
import pandas as pd
```

```python
people1 = {'first' : ['Isaac','Charles','Stephen'],
           'last'  : ['Newton','Darwin','Hawking'],
           'email' : ['Newton@gmail.com','Darwin@gmail.com','Hawking@gmail.com']}
```

```python
df_1 = pd.DataFrame(people1)
df_1
```

```python
people2 = {'first' : ['Thomas','Albert'],
           'last'  : ['Edition','Einstein'],
           'email' : ['Thomas@gmail.com','Einstein@gmail.com']}
```

```python
df_2 = pd.DataFrame(people2)
df_2
```

# Sorting Data

```python
df.sort_values(by = 'last')   # sorted based on last column  will sort ascending order
```

```python
df.sort_values(by = 'last' , ascending = False)    # sorted based on last column ascending = False will sort decending order
```
