### `Apply`:apply is used for calling function on our values adn apply can work on dataframe adn also series object


```python
df['email'].apply(len)    # it return all the len of email
```

```python
def update_email(email):
    return email.upper()     # we can use a simple function we can also use lambda function
```

```python
df['email'].apply(update_email)  # we are getting back of series of email address in upper case
```

```python
df  # it doesn't cahnge in original data 
```

```python
df['email'] = df['email'].apply(update_email)
```

```python
df      # now it will chanhge our original data
```

```python
df['email'] = df['email'].apply(lambda x : x.lower())  # here we have to use lambda function
```

```python
df
```

```python
# now see how apply works on dataframe
# when we run apply on dataframe it runs a function on each row and column of data frames 


df['email'].apply(len)
```

```python
# not apply every value in our dataframe
# it gives no.of rows in each column


df.apply(len)
```

```python
df.apply(len,axis='columns')
```
